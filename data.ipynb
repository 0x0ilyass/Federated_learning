```python
!pip install torch torchvision matplotlib
import matplotlib.pyplot as plt
from torchvision import datasets, transforms
import numpy as np

# Load MNIST dataset
mnist_train = datasets.MNIST(root='./data', train=True, download=True, transform=transforms.ToTensor())

# Select 10 random images and labels
np.random.seed(42)
indices = np.random.choice(len(mnist_train), 10, replace=False)
images = [mnist_train[i][0].squeeze().numpy() for i in indices]
labels = [mnist_train[i][1] for i in indices]

# Prepare data for plotting
fig, axes = plt.subplots(1, 10, figsize=(15, 2))
for i, ax in enumerate(axes):
    ax.imshow(images[i], cmap='gray')
    ax.set_title(f'Label: {labels[i]}')
    ax.axis('off')

plt.tight_layout()
plt.savefig('mnist_sample_images.png')
plt.show()

# Return the image file path for chart creation
'mnist_sample_images.png'
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import fetch_openml

# Chargement du dataset MNIST
mnist = fetch_openml('mnist_784', version=1, as_frame=False)
images = mnist.data.reshape(-1, 28, 28)
labels = mnist.target.astype(int)

# Sélection de 10 images aléatoires
np.random.seed(42)
indices = np.random.choice(len(images), 10, replace=False)
selected_images = images[indices]
selected_labels = labels[indices]

# Affichage
fig, axes = plt.subplots(1, 10, figsize=(15, 2))
for i, ax in enumerate(axes):
    ax.imshow(selected_images[i], cmap='gray')
    ax.set_title(f'Label: {selected_labels[i]}')
    ax.axis('off')
plt.tight_layout()
plt.savefig('mnist_sample_images.png')
plt.show()

# Return the image file path for chart creation
'mnist_sample_images.png'

```
